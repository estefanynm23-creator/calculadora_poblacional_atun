# Exploratory Analysis of Tuna Population Data

This notebook is designed for exploratory data analysis (EDA) of tuna population data. It includes visualizations and preliminary findings to understand the dynamics of tuna populations better.

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set the aesthetic style of the plots
sns.set(style='whitegrid')
```

```python
# Load the sample data
data = pd.read_csv('../src/data/sample_data.csv')

# Display the first few rows of the dataset
data.head()
```

```python
# Summary statistics of the dataset
data.describe()
```

```python
# Visualize the population growth over time
plt.figure(figsize=(12, 6))
plt.plot(data['Year'], data['Population'], marker='o')
plt.title('Tuna Population Growth Over Time')
plt.xlabel('Year')
plt.ylabel('Population')
plt.grid(True)
plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(10, 8))
correlation = data.corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

## Preliminary Findings

- The population of tuna shows a general increasing trend over the years.
- There are significant correlations between various factors affecting tuna population growth.
- Further analysis is needed to understand the underlying causes of population changes.
